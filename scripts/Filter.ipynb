```python
import json
f = open("listSongJSON.json")
data = json.load(f)
data
```

```python
import os
dirlist = [ item for item in os.listdir('.') if os.path.isdir(os.path.join('.', item)) ]
dirlist
```

```python
import shutil
```

```python
n_found = 0
n_missing = 0

for id in data:
    print('-----------------')
    print(f'Processing {id}')

    midi_filename = f'{id}.mid'

    found = False
    path = ''

    for d in dirlist:
        currPath = f'{d}/{midi_filename}'
        if os.path.exists(currPath):
            path = currPath
            found = True
            break
    print(f'found: {found}')
    print(f'path: {path}')

    if found:
        n_found += 1
        shutil.move(path, f'out/{midi_filename}')
    else:
        n_missing += 1

print()
print(f'n_found: {n_found}')
print(f'missing: {n_missing}')
```
